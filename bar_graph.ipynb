```python
import math
from ipycanvas import *
from ipycanvas import Canvas

def data():
    done = False
    data = {}
    while done == False:
        data_name = input("What is the category you will be graphing called? ")
        try:
            data_num = int(input("What is the numerical value of the data collected in this category? "))
        except ValueError:
            print("Invalid input. Please enter a valid number.")
            continue
        data[data_name] = data_num
        is_done = input("Are you finished entering data? (yes or no) ")
        is_done = is_done.lower()
        if is_done == "yes":
            done = True
    return data

def graph(data):
    bar_width = 50
    spacing = 10
    width = len(data) * (bar_width + spacing)
    highest = 0
    for key in data:
        if data[key] > highest:
            highest = data[key]
    height = (highest * 10) + 10
    canvas = Canvas(width=width, height=height)
    canvas.fill_style = "white"
    canvas.fill_rect(0, 0, width, height)
    x = 0
    colors = ["red", "blue", "green", "yellow", "purple", "orange", "pink"]
    color = 0
    for key in data:
        canvas.fill_style = colors[color]
        if color < len(colors) - 1:
            color += 1
        else:
            color = 0
        bar_height = int(data[key]) * 10
        canvas.fill_rect(x, height - bar_height, bar_width, bar_height)
        canvas.stroke_style = "black"
        canvas.stroke_rect(x, height - bar_height, bar_width, bar_height)
        canvas.fill_style = "black"
        length = len(key)
        if length + 5 <= int(data[key]):
            font_size = 16
        else:
            font_size = 12
        canvas.font = f"{int(font_size)}px Arial"
        canvas.text_align = "left"
        canvas.text_baseline = "middle"
        canvas.save()
        bar_center = x + (bar_width / 2)
        canvas.translate(bar_center, height)
        canvas.rotate(-math.pi / 2)
        canvas.fill_text(f"{key} ({data[key]})", 10, 0)
        canvas.restore()
        x += bar_width + spacing
    return canvas

def main():
    graph_data = data()
    final_graph = graph(graph_data)
    display(final_graph)

main()
```
